Faça consultas no dataframe para exibir:
- Nome e time dos jogadores que possuem salário acima de R$ 200000,00
- Nome e salário dos jogadores dos times de Minas Gerais (estado MG)
- Nome e time dos jogadores cujo nome contenha a letra 'u'
- Nome, salário e time dos jogadores, ordenados pelo salário em ordem decrescente
- Nome, salário e time dos jogadores, ordenados pelo nome do time em ordem crescente, e em seguida pelo salário em ordem decrescente
- A quantidade de jogadores por time
- A média salarial por time

```python
#Nome e time dos jogadores que possuem salário acima de R$ 200000,00
resultado = jogadores[jogadores['salario'] > 200000.00][['jogador', 'time']]
print(resultado)
```

```python
#Nome e salário dos jogadores dos times de Minas Gerais (estado MG
mg = jogadores[jogadores['time'] == "Atlético" ]   [['jogador', 'salario']]
mg
```

```python
mg = jogadores[(jogadores['time'] == "Atlético") | (jogadores['time'] == "Cruzeiro")][['jogador', 'salario']]
mg
```

```python
#- Nome e time dos jogadores cujo nome contenha a letra 'u'
jogadores_u = jogadores[jogadores['jogador'].str.contains('u', case=False)][['jogador', 'time']]
print(jogadores_u)
```

```python
#Nome, salário e time dos jogadores, ordenados pelo salário em ordem decrescente
jogadores.sort_values(by=['salario'], ascending = False)[['jogador', 'salario', 'time']]
```

```python
#Nome, salário e time dos jogadores, ordenados pelo nome do time em ordem crescente, e em seguida pelo salário em ordem decrescente
jogadores.sort_values(by=['time', 'salario'], ascending=[True, False])[['jogador', 'salario', 'time']]
```

```python
#A quantidade de jogadores por time
quantidade_jogadores_por_time = jogadores['time'].value_counts()
print(quantidade_jogadores_por_time)
```

```python
#A média salarial por time
media_salarial_por_time = jogadores.groupby('time')['salario'].mean()
print(media_salarial_por_time)
```

```python
#Exporte o conteúdo do dataframe para um novo arquivo CSV, nomeado novo_time.csv
jogadores.to_csv('novo_time.csv', index=False)
```

```python
#Gere um gráfico de barras exibindo a quantidade de jogadores por posição
import matplotlib.pyplot as plt

# Contagem de jogadores por posição
contagem_posicao = jogadores['posicao'].value_counts()

# Criar o gráfico de barras
plt.bar(contagem_posicao.index, contagem_posicao.values)

# Configurar o título e os rótulos dos eixos
plt.title('Quantidade de Jogadores por Posição')
plt.xlabel('Posição')
plt.ylabel('Quantidade de Jogadores')

# Exibir o gráfico
plt.show()
```
